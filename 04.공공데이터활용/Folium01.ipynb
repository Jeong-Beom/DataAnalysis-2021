```python
location = [37.541, 126.986]
folium.Map(location, tiles = 'Stamen Toner', zoom_start = 12)
```

```python
# Markers
map = folium.Map(location, zoom_start = 12)
folium.Marker(
    location = [37.566317, 126.977829], # 위도, 경도
    popup = '서울 지도',
    tooltip = '서울특별시청',
    icon = folium.Icon(color = 'red', info = 'info-sign')
).add_to(map)
folium.Marker(
    location = [37.577317, 126.977829], # 위도, 경도
    popup = 'popup',
    tooltip = 'tooltip'
).add_to(map)
map
```

```python
# Circle Marker
map = folium.Map(location, zoom_start = 12)
folium.Circle(
    radius = 200,
    location = [37.566317, 126.977829], # 위도, 경도
    popup = '서울 지도, popup',
    tooltip = '서울특별시청, tooltip',
    color = 'crimson',
    fill = True
).add_to(map)
folium.CircleMarker(
    radius = 50,
    location = [37.577317, 126.977829], # 위도, 경도
    popup = 'popup',
    tooltip = 'tooltip',
    color = '#3186cc', 
    fill = False
).add_to(map)
map
```
