```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
stop_words.add('said')
```

```python
filename = 'C:\\Workspace\\python\\Data_Science\dataA\\Alice in Wonderland.txt'
with open(filename, 'r', encoding='UTF-8') as f:
    text = f.read()
wc = WordCloud(background_color = 'white',
     max_words = 1000, stopwords = stop_words)
wc.generate(text)
```

```python
keys = list(wc.words_.keys())
values = list(wc.words_.values())
for i in range(10):
    print(keys[i], ':', values[i])
```

```python
plt.figure(figsize = (10, 6))
plt.imshow(wc, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

```python
filename = 'C:\\Workspace\\python\\Data_Science\dataA\\A_new_hope.txt'
filename2 = 'C:\\Workspace\\python\\Data_Science\dataA\\Stormtrooper_mask.png'
with open(filename, 'r', encoding='UTF-8') as f:
    text = f.read()
text = text.replace('HAN', 'Han')
text = text.replace('LUKE\'S', 'LUKE')
from wordcloud import STOPWORDS
stop_words = set(STOPWORDS)
stop_words.add('int')
stop_words.add('ext')
len(stop_words)
```

```python
mask = np.array(Image.open(filename2))
wc = WordCloud(mask = mask, max_words = 1000, stopwords = stop_words,
               margin = 10, random_state = 2021)
wc.generate(text)
default_colors = wc.to_array
```

```python
import random
def grey_color(word, font_size, position,
               orientation, random_state = None, **kwargs):
    return f'hsl(0, 0%, {random.randint(60, 100)}%)'
```

```python
plt.figure(figsize = (12, 12))
plt.imshow(wc.recolor(color_func = grey_color, random_state = 2021),
           interpolation = 'bilinear')
plt.axis('off')
plt.show()
```
