```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fonm
from matplotlib import font_manager, rc
mpl.rcParams['axes.unicode_minus'] = False
font_path = "C:/Windows/Fonts/정직_1.ttf"
font = fonm.FontProperties(fname=font_path).get_name()
rc('font', family=font)
```

```python
pop = pd.read_csv('C:\\Workspace\\python\\Data_Science\dataA\\Cartogram Practice Data\\CCTV\\서울시인구(전처리).csv', encoding = 'UTF-8')
cctv = pd.read_csv('C:\\Workspace\\python\\Data_Science\dataA\\Cartogram Practice Data\\CCTV\\서울시CCTV(전처리).csv', encoding = 'UTF-8')
pop.head()
```

```python
cctv1 = cctv.set_index('자치구')
del cctv1['총계']
cctv1 = cctv1.drop(['계'])
cctv1.T.plot(figsize = (20, 8)) # 특별한 입력을 하지 않아도 DataFrame의 형태에 맞춰 적절하게 그래프를 반환
plt.title('서울시 자치구별 CCTV수 증감현황')
plt.xlabel('연도')
plt.ylabel('CCTV수')
plt.grid()
plt.show()
```

```python
cctv2 = cctv1.reset_index()
cctv2
```

```python
import json
with open('C:\\Workspace\\python\\Data_Science\\dataA\\Cartogram Practice Data\\CCTV\\seoul_municipalities_geo.json', encoding = 'UTF-8') as json_file:
    geo_data = json.load(json_file)
geo_data['features'][0]
```

```python
import folium
center = [37.541, 126.986]
map = folium.Map(location = center, zoom_start = 11)
folium.TileLayer('cartodbpositron').add_to(map)
folium.Choropleth(
    geo_data = geo_data,              # 기존 데이터 파일 경로(인코딩 주의)
    data = cctv2,                        # 시각화하고자하는 데이터프레임
    columns = ('자치구', '2020년'),         # 지도 데이터와 매핑할 값, 시각화하고자하는 변수
    key_on = 'feature.properties.name', # 데이터 파일과 매핑할 값
    fill_color = 'BuPu',              # Color Map
    lengend_name = 'CCTV수'       # Color 범주 이름
).add_to(map)
title = '<h3 align = "center" style = "font-size : 20px"> 서울시 CCTV수(2020년)</h3>'    
map.get_root().html.add_child(folium.Element(title))
map
```
