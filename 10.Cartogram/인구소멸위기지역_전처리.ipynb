```python
df = pd.read_csv('C:\\Workspace\\python\\Data_Science\\dataA\\Cartogram Practice Data\\시군구_성_연령_5세_별_인구2_2020.csv', encoding = 'euc-kr')
print(df.shape)
df.head(270).tail(5)
```

```python
df['시군구'][df.행정구.notnull()] = df.행정구[df.행정구.notnull()]
df.head(266).tail()
```

```python
# 소계 데이터 제외
df = df[df.시군구 != '소계']
print(df.shape)
df.head()
```

```python
df['20~39세'] = df['20 - 24세'] + df['25 - 29세'] + df['30 - 34세'] + df['35 - 39세']
df['65세 이상'] = df['65 - 69세'] + df['70 - 74세'] + df['75 - 79세'] + df['80세 이상']
df = df[['광역시도', '시군구', '구분' ,'인구수', '20~39세', '65세 이상']]
df.head()
```

```python
# pivot_table을 만들어서 시군구 단위로 grouping
pop = pd.pivot_table(df,
                     index = ['광역시도', '시군구'], columns = ['구분'],
                     values =['인구수', '20~39세', '65세 이상'])
print(pop.shape)
pop.head()
```

```python
pop['소멸비율'] = pop['20~39세','여자'] / pop['65세 이상','계']
pop.head()
```

```python
pop['소멸위기지역'] = pop.소멸비율 < 0.5
pop['소멸위기고위험지역'] = pop.소멸비율 < 0.2
pop.head()
```

```python
crisis_region = pop[pop.소멸위기지역].index.get_level_values(1)
len(crisis_region)
print(crisis_region)
```

```python
high_crisis_region = pop[pop.소멸위기고위험지역].index.get_level_values(1)
len(high_crisis_region)
print(high_crisis_region)
```

```python
pop.columns.get_level_values(0)
```

```python
for i in range(len(pop.columns.get_level_values(0))):
    print(pop.columns.get_level_values(0)[i] + pop.columns.get_level_values(1)[i])
```

```python
tmp_col = [pop.columns.get_level_values(0)[i] + \
           pop.columns.get_level_values(1)[i]
           for i in range(len(pop.columns.get_level_values(0)))]
pop.columns = tmp_col
pop.reset_index(inplace = True)
pop['시군구'] = pop.시군구.apply(lambda x: x.strip())
pop.head()
```
