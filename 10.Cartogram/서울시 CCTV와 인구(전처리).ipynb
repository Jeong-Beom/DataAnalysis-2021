```python
import numpy as np
import pandas as pd
```

```python
pop = pd.read_csv('C:\\Workspace\\python\\Data_Science\\dataA\\Cartogram Practice Data\\CCTV\\report.txt', sep = '\s+')
pop.head()
```

```python
pop.columns = ['기간', '자치구', '세대', '합계', '합계(남자)', '합계(여자)', '한국인', '한국인(남자)', '한국인(여자)', '등록외국인', '등록외국인(남자)', \
               '등록외국인(여자)', '세대당인구', '65세이상고령자']
pop.head()
```

```python
pop.drop([0, 1], inplace = True)
```

```python
pop.reset_index(inplace = True)
del pop['index']
pop.head()
```

```python
pop.columns
```

```python
import re
for col_name in pop.columns:
    pop[col_name] = pop[col_name].apply(lambda x: re.sub('[^ㄱ-ㅎㅏ-ㅣ가-힣0-9./]', '', x))
    if col_name != '기간' and col_name != '자치구' and col_name != '세대당인구':
        pop[col_name] =  pop[col_name].astype(float)
        pop[col_name] =  pop[col_name].astype(int)
    elif col_name == '세대당인구':
        pop[col_name] =  pop[col_name].astype(float)
pop.head()
```

```python
pop.head()
```

```python
del pop['기간']
pop.to_csv('C:\\Workspace\\python\\Data_Science\\dataA\\Cartogram Practice Data\\CCTV\\서울시인구(전처리).csv', index = False)
```
