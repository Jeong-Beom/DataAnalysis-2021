```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fonm
from matplotlib import font_manager, rc
mpl.rcParams['axes.unicode_minus'] = False
font_path = "C:/Windows/Fonts/정직_1.ttf"
font = fonm.FontProperties(fname=font_path).get_name()
rc('font', family=font)
```

```python
pop = pd.read_csv('C:\\Workspace\\python\\Data_Science\\dataA\\Cartogram Practice Data\\시군구_ID부여.csv', encoding = 'euc-kr')
pop.head()
```

```python
map_raw = pd.read_excel('C:\\Workspace\\python\\Data_Science\\dataA\\Cartogram Practice Data\\draw_korea_raw(2021).xlsx')
map_raw
```

```python
map = pd.DataFrame(map_raw.stack())
map.reset_index(inplace = True)
print(map.shape)
map.head()
```

```python
map.columns = ['y','x','ID']
map.head()
```

```python
from draw_map import BORDER_LINES, drawKorea, drawKoreaMinus

plt.figure(figsize=(8,11))

for idx, row in map.iterrows():
    # 광역시는 구 이름이 겹치는 경우가 많아서 광역시 이름도 같이 표시 
    # (중구, 서구 등)
    if len(row.ID.split())==2:
        dispname = f'{row.ID.split()[0]}\n{row.ID.split()[1]}'
    elif row.ID[:2]=='고성':
        dispname = '고성'
    else:
        dispname = row.ID

    # 서대문구, 서귀포시 같이 이름이 3자 이상인 경우에 작은 글자로 표시
    if len(dispname.splitlines()[-1]) >= 3:
        fontsize, linespacing = 9.5, 1.5
    else:
        fontsize, linespacing = 11, 1.2

    plt.annotate(dispname, (row['x']+0.5, row['y']+0.5), weight='bold',
                 fontsize=fontsize, ha='center', va='center', 
                 linespacing=linespacing)
    
# 시도 경계
for path in BORDER_LINES:
    ys, xs = zip(*path)
    plt.plot(xs, ys, c='black', lw=1.5)

plt.gca().invert_yaxis()
#plt.gca().set_aspect(1)

plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import json
with open('C:\\Workspace\\python\\Data_Science\\dataA\\Cartogram Practice Data\\skorea_municipalities_geo_simple.json', encoding = 'UTF-8') as json_file:
    geo_data = json.load(json_file)

map = folium.Map(location=[35.9902, 127.054], zoom_start=7)
folium.Choropleth(
    geo_data=geo_data,
    data=pop.인구수계,
    columns=(pop.index, pop.인구수계),
    key_on='feature.id',
    fill_color='YlGnBu',
    legend_name='인구수'
).add_to(map)

title = '<h3 align="center" style="font-size:20px">기초자치단체별 인구수</h3>'
map.get_root().html.add_child(folium.Element(title))
map
```
