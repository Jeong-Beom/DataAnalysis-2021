```python
# 식신 / 명칭, 주요메뉴, 평점, 전화번호, 주소
import requests
import pandas as pd
from urllib.parse import quote

# 데이터 가져오기
base_url = 'https://www.siksinhot.com'
url = f'{base_url}/search?keywords={quote("홍대입구")}'
req = requests.get(url)
html = req.text
# html
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')
# 찾으려고하는 데이터의 태그 찾기
lis = soup.select('div.listTy1 > ul > li')
print(len(lis))

li = lis[0]
href = li.select_one('a')['href']
sub_url = base_url + href
req = requests.get(sub_url)
sub_soup = BeautifulSoup(req.text, 'html.parser') 
```

```python
store = sub_soup.select_one('.title')
info = store.find('h3').get_text()
score = store.select_one('h3 > strong').string # 평점
if score == None:
    score = 0.0
    name = info[:info.find('평가중')]
else:
    score = store.select_one('h3 > strong').string
    name = info[:info.find(score)]
menu = sub_soup.select('.store_info p')[1].get_text() #  메뉴
tel = sub_soup.select_one('.p_tel p').get_text()
tel # 전화번호
addr = sub_soup.select_one('.txt_adr').string
addr # 주소
```

```python
name_list, score_list, menu_list, tel_list, addr_list = [], [], [], [], []
for li in lis:
    href = li.select_one('a')['href']
    sub_url = base_url + href
    req = requests.get(sub_url)
    sub_soup = BeautifulSoup(req.text, 'html.parser') 
    store = sub_soup.select_one('.title')
    info = store.find('h3').get_text()
    score = store.select_one('h3 > strong').string # 평점
    if score == None:
        score = '평가중'
        name = info[:info.find('평가중')]
    else:
        score = store.select_one('h3 > strong').string
        name = info[:info.find(score)]
    menu = sub_soup.select('.store_info p')[1].get_text() 
    tel = sub_soup.select_one('.p_tel p').get_text()
    addr = sub_soup.select_one('.txt_adr').string
    name_list.append(name) 
    score_list.append(score)
    menu_list.append(menu)
    tel_list.append(tel)
    addr_list.append(addr)
df = pd.DataFrame({
    '가게이름' : name_list, '평점' : score_list, '메뉴' : menu_list, '전화번호' : tel_list, '주소' : addr_list
})
df 
```
