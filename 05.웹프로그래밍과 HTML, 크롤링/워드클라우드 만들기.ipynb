```python
# 해시태그 데이터로 워드클라우드 만들기
# 한글로 만드려면 한글 형태소 분석기가 필요 - KoNLPy
import pandas as pd
df = pd.read_csv('C:\\Workspace\\python\\Data_Science\\04.공공데이터활용\\data\\제주 핫플레이스.csv')
df.head(5)
df.tags[: 3]
tags_total = []
tags = df.tags[0]
# print(tags[2: -2].split("', '")[: 5])
tags_total = ['시작']
for tags in df.tags:
    tags_total.extend(tags[2: -2].split("', "))
```

```python
# 해시태그 출현빈도
# collections 모듈
from collections import Counter, OrderedDict 
# Counter는 해당 리스트의 중복되는 값을 세는데 사용 / OrderedDict는 순서를 가지지않는 dictionary타입에 순서를 부여
s = 'A quick nrown fox jumps over the lazy dog.'
Counter(s)
sl = ['A', 'a', 'quick', 'brown', 'brown']
Counter(sl)
od = dict()
od['x'] = 1
od['y'] = 2
od['z'] = 3
```

```python
Counter(tags_total)
# 데이터 정제하기 - stopwords
stopwords = ['#jeju', '#일상', '#제주', '#제주도', '#선팔', '#제주눈썹문신', '\'제주'
              '#서귀포눈썹문신', '#반영구', '#눈썹문신', '#제주자연눈썹', '\'#삼육오빠']
tag_total_refined = []
for tag in tags_total:
    if tag not in stopwords:
        tag_total_refined.append(tag)
Counter(tag_total_refined)
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fonm
from matplotlib import font_manager, rc
mpl.rcParams['axes.unicode_minus'] = False
font_path = "C:/Windows/Fonts/정직_1.ttf"
font = fonm.FontProperties(fname=font_path).get_name()
rc('font', family=font)
from wordcloud import WordCloud
wordcloud = WordCloud(font_path = font_path,
                      background_color = 'white',
                      max_words = 100,
                      relative_scaling = 0.3,
                      width = 800, height = 400).generate_from_frequencies(Counter(tag_total_refined))
plt.figure(figsize = (15, 10))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
