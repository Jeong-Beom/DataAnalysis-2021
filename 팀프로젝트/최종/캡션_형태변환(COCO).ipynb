```python
import json
import numpy as np
import pandas as pd

with open('C:/Users/Bestc/Desktop/빅데이터 지능형 서비스 개발과정(멀티캠퍼스)/팀프로젝트/최종/자료/한국어 영상 캡셔닝 데이터셋/MSCOCO_train_val_Korean.json', 'r') as f:
    json_data = json.load(f)

len(json_data),type(json_data)
```

```python
val_list = []
train_list = []

for i in range(len(json_data)):
    if 'val' in json_data[i]['file_path']:
        val_list.append(json_data[i])
    if 'train' in json_data[i]['file_path']:
        train_list.append(json_data[i])
```

```python
file_path1 = 'C:/Users/Bestc/Desktop/빅데이터 지능형 서비스 개발과정(멀티캠퍼스)/팀프로젝트/최종/자료/한국어 영상 캡셔닝 데이터셋/captions_val2014(ko).json'
with open(file_path1, 'w') as outfile:
    json.dump(val_list, outfile)

file_path2 = 'C:/Users/Bestc/Desktop/빅데이터 지능형 서비스 개발과정(멀티캠퍼스)/팀프로젝트/최종/자료/한국어 영상 캡셔닝 데이터셋/captions_train2014(ko).json'
with open(file_path2, 'w') as outfile:
    json.dump(train_list, outfile)
```
