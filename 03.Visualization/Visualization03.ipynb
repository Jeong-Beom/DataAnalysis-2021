```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fonm
from matplotlib import font_manager, rc
mpl.rcParams['axes.unicode_minus'] = False
font_path = 'C:/Windows/Fonts/정직_1.ttf'
font = fonm.FontProperties(fname=font_path).get_name()
rc('font', family=font)

# Pandas를 활용한 데이터 시각화
import seaborn as sns
import pandas as pd
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')

np.random.seed(0)
df1 = pd.DataFrame(np.random.randn(100, 3),
                   index=pd.date_range('1/1/2018', periods=100),
                   columns=['A', 'B', 'C']).cumsum()
df1.tail()
```

```python
df1.plot() # 특별한 입력을 하지 않아도 DataFrame의 형태에 맞춰 적절하게 그래프를 반환
plt.title('Pandas의 Plot메소드 사용 예시')
plt.xlabel('일자')
plt.ylabel('Data')
plt.grid()
plt.show()
```

```python
iris.sepal_length[:20].plot(kind='bar', rot=0, color = plt.get_cmap('RdBu')(np.linspace(0.8, 0.2, 20)))
plt.title("꽃받침의 길이 시각화")
plt.xlabel("Data")
plt.ylabel("꽃받침의 길이")
plt.show()
```

```python
iris[:5].plot.bar(rot = 0) # kind를 직접 명령안해줘도 차트의 종류 지정 가능
plt.ylim(0, 7)
plt.show()
```

```python
df2 = iris.groupby(iris.species).mean()
df2.columns.name = 'feature'
df2.plot.bar(rot = 0)
plt.ylim(0, 8) # y축의 범위 지정
plt.show()
```

```python
df3 = titanic['class'].value_counts()
df3.plot.pie(autopct = '%.2f%%')
plt.title('선실별 승객수 비율')
plt.axis('equal')
plt.show()
```

```python
df2.plot.bar(rot=0)
plt.title("각 종의 Feature별 평균")
plt.xlabel("평균")
plt.ylabel("종")
plt.ylim(0, 8)
plt.show()
```

```python
iris.plot.box()
plt.title("각 Feature 값들의 빈도수에 대한 Box Plot")
plt.xlabel("Feature")
plt.ylabel("데이터 값")
plt.show()
```

```python
iris.boxplot(by = 'species')
plt.tight_layout(pad = 2, h_pad = 0)
plt.show()
```
