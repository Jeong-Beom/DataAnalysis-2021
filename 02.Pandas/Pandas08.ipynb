```python
# 그룹분석
import numpy as np
import pandas as pd
import seaborn as sns

data = {
    "도시": ["서울", "서울", "서울", "부산", "부산", "부산", "인천", "인천"],
    "연도": ["2015", "2010", "2005", "2015", "2010", "2005", "2015", "2010"],
    "인구": [9904312, 9631482, 9762546, 3448737, 3393191, 3512547, 2890451, 263203],
    "지역": ["수도권", "수도권", "수도권", "경상권", "경상권", "경상권", "수도권", "수도권"]
}
columns = ["도시", "연도", "인구", "지역"]
df1 = pd.DataFrame(data, columns=columns)
df1.pivot('도시', '연도', '인구') # 행 인덱스, 열 인덱스, 찾을 데이터
df1.pivot(['지역','도시'], '연도', '인구')

# 그룹분석
iris = sns.load_dataset('iris')
iris.groupby(iris.species).mean()
iris.groupby(iris.species).agg(['mean', 'std'])
np.random.seed(0)
df2 = pd.DataFrame({
    'key1': ['A', 'A', 'B', 'B', 'A'],
    'key2': ['one', 'two', 'one', 'two', 'one'],
    'data1': [1, 2, 3, 4, 5],
    'data2': [10, 20, 30, 40, 50]
})
groups = df2.groupby(df2.key1) #(데이터, 그룹을 나눌 기준요인)
df2.data1.groupby(df2.key1).sum()
groups.sum()
df2.data1.groupby([df2.key1, df2.key2]).sum()
mpg = sns.load_dataset('mpg')
mpg.groupby(mpg.name).sum()
mpg['manufacturer'] = mpg.name.apply(lambda x: x.split()[0])
mpg['model'] = mpg.name.apply(lambda x: ' '.join(x.split()[1:])) 
mpg.head()
```

```python
tips = sns.load_dataset("tips")
tips.tail()
tips['tip_pct'] = tips['tip'] / tips['total_bill']
tips.tail()
tips.groupby('sex')[['tip_pct']].mean() # 성별에 따른 평균 팁 비율
tips.groupby('smoker')[['tip_pct']].mean() # 흡연유무에 따른 평균 팁 비율
tips.groupby(['sex', 'smoker'])[['tip_pct']].mean()
tips.pivot_table('tip_pct', ['sex', 'smoker']) # 찾을 데이터, 행 인덱스, 열 인덱스
tips.groupby('day')[['tip_pct']].mean()
tips.groupby('time')[['tip_pct']].mean()
tips.groupby(['day', 'time'])[['tip_pct']].mean()
tips.pivot_table('tip_pct',['day', 'time'])
# 요일, 시간, 인원수
tips.pivot_table('tip_pct', ['day', 'time'], 'size')
# 어떤 요인이 가장 크게 작용하는지 판단
tips.groupby(['day', 'time'])[['tip_pct']].agg(lambda x: x.max() / x.min())
```
